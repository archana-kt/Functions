# Using List

```python
def getdata(*n):
    x = list()
    for i in n:
        x.append(i)
    print(x)
```

```python
getdata(1,2,3,4,5,6,7,8,9)
```

```python
def getdata(*n):
    x = list()
    for i in n:
        x.append(i)
    return(x)
```

```python
getdata(1,2,3,4,5,6,7,8,9)
```

# Using Dict

```python
def getdata(*n):
    x = dict()
    for i in n:
        x.update(i)
    return(x)
```

```python
getdata({"State":"TN","Country":"India"})
```
